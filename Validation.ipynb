```python
import pickle 
def save_obj(obj, name ):
    with open(name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name ):
    with open(name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
import pandas as pd
import skimage.io as skio
data = pd.read_csv("./val/val_annotations.txt", sep="\t", header=None)
data.columns = ["a","f","b","c","d","e"]

label_dict = load_obj("label_dict")

for index, row in data.iterrows():

    temp_img= skio.imread("./val/images/"+row["a"])
    temp_img= temp_img[int(row["c"]):int(row["e"]),int(row["b"]):int(row["d"])]
    print(temp_img.shape)

    try:
        temp_img=skio.transform.resize(temp_img,(64,64))
    except:
        temp_img= skio.imread("./val/"+"images/"+row["a"])

#     try:
    print(label_dict[row["f"]])
    z="diff_orient_images/img_"+str(label_dict[row["f"]])+"_"+row["a"]
    skio.imsave("./val/"+z,temp_img)
#     except :
#         print("./val/"+"diff_orient_images/"+row["a"])
#     del temp_img
```

```python
import glob 
def prep_val_array():
    test_arr=[]
    test_labels = []
    for j in glob.glob("./val/"+"diff_orient_images/*.JPEG"):
        temp_img= skio.imread(j)
        if len(temp_img.shape)==3:
            test_arr.append(temp_img)
            test_labels.append(int(j.split('/')[3].split('_')[1]))
    return test_arr,test_labels

temp = prep_val_array()
```
