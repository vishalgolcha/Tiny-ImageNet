```python
def prep_train_array(x,fold_dict_x,label_dict):
    train_arr=[]
    train_labels_arr = []
    for i in x:
        for j in glob.glob(fold_dict_x[i]+"diff_orient_images/*.JPEG"):
            temp_img= skio.io.imread(j)
            train_arr.append(temp_img)
            train_labels_arr.append(label_dict[fold_dict_x[i].split('/')[2]])
    return train_arr,train_labels_arr


train_data = prep_train_array(x,fold,label_dict)
```

```python
def save_obj(obj, name ):
    with open(name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name ):
    with open(name + '.pkl', 'rb') as f:
        return pickle.load(f)
```
